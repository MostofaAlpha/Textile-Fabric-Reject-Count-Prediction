```python
import pandas as pd
import numpy as np
import seaborn as sns
import matplotlib.pyplot as plt
from scipy.stats import pearsonr
from math import sqrt

import os #provides functions for interacting with the operating system
import numpy as np 
import pandas as pd
from matplotlib import pyplot as plt
import seaborn as sns
from sklearn.linear_model import LinearRegression
from sklearn.tree import DecisionTreeClassifier
from sklearn.model_selection import train_test_split
from sklearn.metrics import r2_score, explained_variance_score, mean_absolute_error, mean_squared_error
from math import sqrt


from sklearn.model_selection import train_test_split
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score
from sklearn.ensemble import RandomForestRegressor
from sklearn.neighbors import KNeighborsRegressor
from sklearn.model_selection import cross_val_score
from sklearn.ensemble import GradientBoostingRegressor
```

- Delete the **Construction column** because its contents are represented by the warp_count, weft_count, epi, ppi columns
- Delete the **Previous_pdn column** because it contains one value among all records

```python
# del weaving_data['Previous_pdn']

single_value_column = [column for column in weaving_data.columns if weaving_data[column].nunique()==1]

print(f"Column Including Same Value - {single_value_column}")
```

```python
weaving_data = weaving_data.drop(columns=['Construction', *single_value_column])
weaving_data
```

```python
weaving_data.head(5)
```

```python
def remove_sign_type(val):
    if type(val) == str and val.startswith('double_'):
        return val.replace('double_', '')
    return val
```

```python
weaving_data['warp_count'] = weaving_data['warp_count'].apply(remove_sign_type)
weaving_data['warp_count'] = weaving_data['warp_count'].astype(float)
```

# Linear Regression

```python
X_train, X_test, y_train, y_test = train_test_split(weaving_data.drop(columns=["Rejection"]), weaving_data["Rejection"], test_size=0.20, random_state=123)

lm = LinearRegression()

lm.fit(X_train,y_train)

y_pred = lm.predict(X_train)



# Model Accuracy on training dataset

print('The Accuracy  on the training dataset is: ', lm.score(X_train, y_train) )
print('The Accuracy r2  on the training dataset prediction is: ',r2_score(y_train,y_pred) )   

print("")
# Model Accuracy on testing dataset
print('The Accuracy  on the testing dataset is: ', lm.score(X_test, y_test) )


rmse_lm_train = sqrt(mean_squared_error(y_train,y_pred))
rmse_lm_test = sqrt(mean_squared_error(y_test,lm.predict(X_test)))


print("")
# The Mean Squared Error (MSE)
print('The RMSE  on the training dataset is: ',sqrt(mean_squared_error(y_train,y_pred)))
print('The RMSE  on the testing dataset is: ',sqrt(mean_squared_error(y_test,lm.predict(X_test))))

print("")
# The Mean Absolute Error (MAE)
print('The MAE  on the training dataset is: ',mean_absolute_error(y_train,y_pred))
print('The MAE  on the testing dataset is: ',mean_absolute_error(y_test,lm.predict(X_test)))


print("")
# Coefficients
print('Coefficients: ', lm.coef_ )

print("")
# The Intercept
print('Intercept: ', lm.intercept_)
```

# Random Forest

```python
X_train, X_test, y_train, y_test = train_test_split(weaving_data.drop(columns=["Rejection"]), weaving_data["Rejection"], test_size=0.20, random_state=123)

random_forest = RandomForestRegressor(n_estimators=100, random_state=123)

random_forest.fit(X_train, y_train)

y_pred = random_forest.predict(X_train)

# Model Accuracy on training dataset

print('The Accuracy  on the training dataset is: ', random_forest.score(X_train, y_train) )
print('The Accuracy r2  on the training dataset prediction is: ',r2_score(y_train,y_pred) )   

print("")
# Model Accuracy on testing dataset
print('The Accuracy  on the testing dataset is: ', random_forest.score(X_test, y_test) )

rmse_rf_train = sqrt(mean_squared_error(y_train,y_pred))
rmse_rf_test = sqrt(mean_squared_error(y_test,random_forest.predict(X_test)))


print("")
# The Mean Squared Error (MSE)
print('The RMSE  on the training dataset is: ',sqrt(mean_squared_error(y_train,y_pred)))
print('The RMSE  on the testing dataset is: ',sqrt(mean_squared_error(y_test,random_forest.predict(X_test))))

print("")
# The Mean Absolute Error (MAE)
print('The MAE  on the training dataset is: ',mean_absolute_error(y_train,y_pred))
print('The MAE  on the testing dataset is: ',mean_absolute_error(y_test,random_forest.predict(X_test)))


```

# KNN

```python
X_train, X_test, y_train, y_test = train_test_split(weaving_data.drop(columns=["Rejection"]), weaving_data["Rejection"], test_size=0.2)

knn = KNeighborsRegressor(n_neighbors=6)

knn.fit(X_train, y_train)

y_pred = knn.predict(X_train)

# Model Accuracy on training dataset

print('The Accuracy  on the training dataset is: ', knn.score(X_train, y_train) )
print('The Accuracy r2  on the training dataset prediction is: ',r2_score(y_train,y_pred) )   

print("")
# Model Accuracy on testing dataset
print('The Accuracy  on the testing dataset is: ', knn.score(X_test, y_test) )

rmse_knn_train = sqrt(mean_squared_error(y_train,y_pred))
rmse_knn_test = sqrt(mean_squared_error(y_test,knn.predict(X_test)))

print("")
# The Mean Squared Error (MSE)
print('The RMSE  on the training dataset is: ',sqrt(mean_squared_error(y_train,y_pred)))
print('The RMSE  on the testing dataset is: ',sqrt(mean_squared_error(y_test,knn.predict(X_test))))

print("")
# The Mean Absolute Error (MAE)
print('The MAE  on the training dataset is: ',mean_absolute_error(y_train,y_pred))
print('The MAE  on the testing dataset is: ',mean_absolute_error(y_test,knn.predict(X_test)))
```

# Gradient Boosting Regressor

```python
X_train, X_test, y_train, y_test = train_test_split(weaving_data.drop(columns=["Rejection"]), weaving_data["Rejection"], test_size=0.2, random_state=123)

xgbr = GradientBoostingRegressor(n_estimators=500,random_state=123)

xgbr.fit(X_train, y_train)

y_pred = xgbr.predict(X_train)

# Model Accuracy on training dataset

print('The Accuracy  on the training dataset is: ', xgbr.score(X_train, y_train) )
print('The Accuracy r2  on the training dataset prediction is: ',r2_score(y_train,y_pred))   

print("")
# Model Accuracy on testing dataset
print('The Accuracy  on the testing dataset is: ', xgbr.score(X_test, y_test) )

rmse_boosting_train = sqrt(mean_squared_error(y_train,y_pred))
rmse_boosting_test = sqrt(mean_squared_error(y_test,xgbr.predict(X_test)))

print("")
# The Root Mean Squared Error (MSE)
print('The RMSE  on the training dataset is: ',sqrt(mean_squared_error(y_train,y_pred)))
print('The RMSE  on the testing dataset is: ',sqrt(mean_squared_error(y_test,xgbr.predict(X_test))))

print("")
# The Mean Absolute Error (MAE)
print('The MAE  on the training dataset is: ',mean_absolute_error(y_train,y_pred))
print('The MAE  on the testing dataset is: ',mean_absolute_error(y_test,xgbr.predict(X_test)))
```

```python
from sklearn.model_selection import GridSearchCV
X_train, X_test, y_train, y_test = train_test_split(weaving_data.drop(columns=["Rejection"]), weaving_data["Rejection"], test_size=0.2, random_state=123)



xgbr = XGBRegressor()

param_grid = {
           'max_depth': [3,6,9,12],
           'learning_rate': [0.01, 0.05, 0.1],
           'n_estimators': [100, 500, 1000],
           'colsample_bytree': [0.3, 0.7]
}

grid_search = GridSearchCV(estimator = xgbr, param_grid = param_grid, 
                          cv = 3, n_jobs = -1, verbose = 2)

grid_search.fit(X_train, y_train)

grid_search.best_params_
```
